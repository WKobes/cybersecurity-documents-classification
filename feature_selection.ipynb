```python
import os
import json
import re
```

```python
relevant_dir = os.path.abspath(os.curdir) + '/training/relevant/'
not_relevant_dir = os.path.abspath(os.curdir) + '/training/irrelevant/'

######################## Relevant

word_counts = []
document_types = {}
tmp = []
years = []


print("Relevant")

# Loop through all relevant
for filename in os.listdir(relevant_dir):
    
    # Skip directories
    if os.path.isdir(os.path.join(relevant_dir, filename)):
        continue
    
    
    # Open every file
    with open(relevant_dir + filename, 'r') as file:
        file_data = json.loads(file.read())
        
        
        word_counts.append(file_data['word_counts'].values())
        tmp = list(file_data['word_counts'].keys())
        
        years += [int(re.search(r"[0-9]{4}", file_data['document_date']).group(0))]
        
        current_type = file_data['document_type']
        if current_type:
            document_types[current_type] = document_types.get(current_type, 0) + 1
        else:
            document_types['Unknown'] = document_types.get('Unknown', 0) + 1

averages = [sum(col) / float(len(col)) for col in zip(*word_counts)]
relevant = dict(zip(tmp,averages))

print(sum(years)/len(years))

# print(json.dumps(document_types, indent=2))

print(json.dumps(relevant, indent=2))


######################### Irrelevant

print("Irrelevant")

word_counts = []
document_types = {}
tmp = []
years = []

# Loop through all irrelevant
for filename in os.listdir(not_relevant_dir):
    
    # Skip directories
    if os.path.isdir(os.path.join(not_relevant_dir, filename)):
        continue
    
    # Open every file
    with open(not_relevant_dir + filename, 'r') as file:
        file_data = json.loads(file.read())
        word_counts.append(file_data['word_counts'].values())
        tmp = list(file_data['word_counts'].keys())
        current_type = file_data['document_type']
        
        years += [int(re.search(r"[0-9]{4}", file_data['document_date']).group(0))]
        
        if current_type:
            document_types[current_type] = document_types.get(current_type, 0) + 1
        else:
            document_types['Unknown'] = document_types.get('Unknown', 0) + 1

averages = [sum(col) / float(len(col)) for col in zip(*word_counts)]

not_relevant = dict(zip(tmp,averages))

super_dict = dict()

for rel in relevant.keys():
    if not_relevant[rel] > 0:
        super_dict[rel] = relevant[rel] / not_relevant[rel]
    else:
        super_dict[rel] = relevant[rel]


# print(json.dumps(document_types, indent=2))

print(sum(years)/len(years))
print(json.dumps(not_relevant, indent=2))  

print("Other")

print(json.dumps(super_dict, indent=2))
```
